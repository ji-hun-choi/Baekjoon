## 접근방법

- for 문으로 n개를 입력 받아 리스트에 저장
- 다시 for 문으로 1~n까지 append 하다가 앞에 받아놓은 수를 만나면 while로 반복 하며 pop() 없으면 break


```python
import sys
rl = sys.stdin.readline
n = int(rl())
count = 1
stack = []
result = []
for i in range(1, n + 1): # 데이터 개수만큼 반복
    data = int(rl())
    while count <= data: # 입력 받은 데이터에 도달할 때까지 삽입
        stack.append(count)
        count += 1
        result.append('+')
    if stack[-1] == data: # 스택의 최상위 원소가 데이터와 같을 때 출력
        stack.pop()
        result.append('-')
    else: # 불가능한 경우
        print('NO')
        exit(0)   # NO 출력 후 프로세스 종료!!
print(*result, sep='\n') # 가능한 경우  #
```
