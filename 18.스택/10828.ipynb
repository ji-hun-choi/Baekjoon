## 접근방법

- 처음에 숫자를 입력받는다
- push, size, empty, pop, top을 str로 받아오면 처리한다.
- push 일 경우 뒤는 숫자로 처리한다.
- stack 처리는 리스트로 진행

```python
import sys
rl = sys.stdin.readline

num = int(rl())
stack = []

for _ in range(num):
    a = rl().split()
    order = a[0]
    if order == 'push':stack.append(int(a[1]))
    elif order == 'top':
        if len(stack) == 0:print(-1)
        else : print(stack[-1])
    elif order == 'size':print(len(stack))
    elif order == 'pop':
        if len(stack) == 0:print(-1)
        else :print(stack.pop())
    elif order == 'empty':
        if len(stack) == 0: print(1)
        else : print(0)
```
