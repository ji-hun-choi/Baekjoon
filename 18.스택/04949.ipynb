## 접근방법

- 이전 문제에 사용했던 idx_c 와 check 함수를 재 사용한다.
- idx_c 에서 조건에 '[' 도 추가를 해주며 idx2를 추가해 '['에 대한 갯수를 출력한다.
- 문자가 '.' 만 나오면 종료 되는 식으로 무한 while문을 작성한다.
- '(space).' 도 yes가 나와야 된다.

```python
def idx_c (x):
    idx1 = 0
    idx2 = 0
    for i in x:
        if idx1 <= -1 or idx2<= -1:
            return idx1, idx2
        elif i == '(' :
            idx1 += 1
        elif i == '[' :
            idx2 += 1
        elif i == ')' :
            idx1 -= 1
        elif i == ']':
            idx2 -= 1
    return idx1, idx2

def cheack(idx1, idx2):
    if idx1 == 0 and idx2 == 0 :
        print('yes')
    else :
        print('no')

while True:
    x = input()
    if x is '.':
        break
    else :
        a, b = idx_c(x)
        cheack(a, b)
```

- 문제에서는 '([)]'  도 허락을 안해준 것 같다.
- stack 형으로 넣고 빼고 해본다.

```python
def isEmpty(x):
    if len(x) == 0:
        return True
    else :
        return False

def check_string(x):
    lis1 = []
    for i in x:
        if i == '(':
            lis1.append(i)
        elif i == '[':
            lis1.append(i)
        elif i == ')':
            if not isEmpty(lis1):
                str1 = lis1.pop()
                if str1 != '(':
                    return 'no'
            else :
                return 'no'
        elif i == ']':
            if not isEmpty(lis1):
                str1 = lis1.pop()
                if str1 != '[':
                    return 'no'
            else :
                return 'no'
    if isEmpty(lis1):
        return 'yes'
    else :
        return 'no'

while True:
    x = input()
    if x == '.':
        break
    else :
        string = check_string(x)
        print(string)
```
