## 접근방법

- 괄호가 '('(추가) 되면 리스트에 추가를 한다.
- ')'(pop)할때 리스트 길이를 체크를 한다
- 길이가 0인데 pop을 시도하면 NO를 출력 후 break 

```python
import sys
rl = sys.stdin.readline

num = int(input())


def isEmpty(lis1):
    if len(lis1)==0:
        return True
    else :
        return False

for _ in range(num):
    lis1 = []
    parents = input()
    for parent in parents:
        if parent == '(': # '(' 이면 lis1 에 추가
            lis1.append(parent)
        else :
            if not isEmpty(lis1): # 비어있지 않으면 빼기
                lis1.pop()
            else:
                lis1.append(parent) # 아래 정답이 나오기 위해 비어있는곳에 ')' 입력 후 for문 종료
                break
    if isEmpty(lis1): # 비어 있으면 YES
        print('YES')
    else:             # 안에 있으면 NO
        print('NO')
```

- 예제는 잘 나오지만 정답이 아니다.


- 인덱스로 접근을 했다.
> **idx_**
- '(' 일때 idx += 1
- ')' 일때 idx -= 1
- idx <= -1 return -1
> **cheacking**
- x == 0 일때 YES
- else 일때 NO

```python
import sys
rl = sys.stdin.readline

def idx_c (x):
    idx = 0
    for i in x:
        if idx <= -1:
            return idx
        elif i == '(':
            idx += 1
        elif i == ')':
            idx -= 1
    return idx

def cheack(x):
    if x == 0 :
        print('YES')
    else :
        print('NO')

num = int(input())
for i in range(num):
    x = input()
    cheack(idx_c(x))
```
