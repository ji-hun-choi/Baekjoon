## 접근방법

- 입력을 리스트에 모두 받는다
- 답 리스트를 -1 로 num 의 크기만큼 초기화 해준다.
- 이중 for문으로 리스트를 돌리는데 인덱스로 처리한다.
- 첫번째 while문에서는 idx 로 비교 대상을 찾는다.
- 두번째 for문에서는 count+1 ~ 입력리스트 길이 만큼 다음 숫자를 현재 수와 크기를 비교한다
- 만약 다음 숫자가 크면 해당 숫자를 출력하고 계속 진행하며 크기가 크면 answer에 넣어주고 break한다.

```python
import sys

rl = sys.stdin.readline

# num = int(input())
num = int(rl())
answer = [-1] * num
idx = 0


# quest = list(map(int, input().split(' ')))
quest = list(map(int, rl().split(' ')))

while idx < num:
    for i in range(idx+1, len(quest)):
        if quest[idx] < quest[i]:
            answer[idx] = quest[i]
            break
    idx += 1
print(*answer)

```

- 시간 초가 에러가 나온다
- 아마 큰수가 나올때 까지 계속 찾는것 때문에 그런것 같다
- 이번에는 스택형으로 만들어서 했다

```python
import sys
n = int(sys.stdin.readline())
lis = list(map(int, sys.stdin.readline().split()))
ans = [-1] * n
index = []
for i in range(0, n):
    while len(index) != 0 and lis[index[-1]] < lis[i]:
        ans[index.pop()] = lis[i]
    index.append(i)

print(*ans)
```
