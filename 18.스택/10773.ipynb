## 접근방법

- 첫번째 정수가 줄을 정한다.
- 0~100000 사이 값 이다

- 아닐경우 해당 수를 쓴다.
- 0을 쓰기 위해서는 지울 수 있는 수가 있음을 보장.
- 리스트를 생성하여 스택 활용

```python
k = int(input())
lis1 = list()

for _ in range(k):
    n = int(input())
    n_b = 0
    if n > 0 :
        lis1.append(n)
    else :
        lis1.pop()
print(sum(lis1))
```
